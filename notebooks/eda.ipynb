```python
## comprobación de nuevos datos 
table_name = 'Almacenes U6321303'
column_name = 'Codi'

# Paso 1: Consultar los valores actuales en la base de datos
query = f'SELECT "{column_name}" FROM \"{table_name}\"'
existing_values = pd.read_sql(query, con=engine)[column_name].tolist()
```

```python
existing_values
```

```python
# Paso 2: Obtener los valores de la columna 'Codi' en el DataFrame
df_codi_values = df_almacenes['Codi'].astype(str).tolist()
```

```python
# Paso 3: Comparar los valores para identificar los nuevos
new_values = set(df_codi_values) - set(existing_values)
```

```python
df_codi_values
```

```python
existing_values
```

```python
set(df_codi_values) - set(existing_values)
```

```python
# Paso 4: Insertar los valores nuevos en la base de datos
if new_values:
    # Filtrar el DataFrame para obtener solo las filas con nuevos valores de 'Codi'
    df_new_data = df_almacenes[df_almacenes[column_name].isin(new_values)]
    
    # Insertar los nuevos datos en la tabla
    df_new_data.to_sql(table_name, con=engine, if_exists='append', index=False)
    print(f"{len(df_new_data)} nuevos registros insertados en '{table_name}'.")
else:
    print(f"No hay nuevos datos que ingestar en '{table_name}'.")
```
